# Network analysis

```python
import networkx as nx
import matplotlib.pyplot as plt

# Create a graph
G = nx.Graph()
G.clear()
# Iterate through the dataset and add nodes and edges for matches
for index, row in df.iterrows():
    if row['match'] == 1:
        G.add_node(int(row['iid']), gender=row['gender'], age=row['age'])
        G.add_node(int(row['pid']), gender=row['gender'], age=row['age'])
        partner_id = int(row['pid'])
        G.add_edge(int(row['iid']), partner_id)

# remove nodes with no edges
G.remove_nodes_from(list(nx.isolates(G)))

# Visualize the network
pos = nx.spring_layout(G)
nx.draw(G, pos, node_size=20)
plt.title("Speed Dating Network")
plt.show()
```

```python
import networkx as nx
import matplotlib.pyplot as plt

# Create a graph
G = nx.Graph()

# Create a dictionary to map gender to node colors
gender_colors = {0: 'blue', 1: 'pink'}

# Iterate through the dataset and add nodes and edges for matches
for index, row in df.iterrows():
    if row['match'] == 1:
        G.add_node(int(row['iid']), gender=row['gender'], age=row['age'], color=gender_colors[row['gender']])
        G.add_node(int(row['pid']), gender=row['gender'], age=row['age'], color=gender_colors[row['gender']])
        partner_id = int(row['pid'])
        G.add_edge(int(row['iid']), partner_id)

# Remove isolated nodes
isolated_nodes = list(nx.isolates(G))
G.remove_nodes_from(isolated_nodes)

# Extract node colors from the 'color' attribute
node_colors = [data['color'] for node, data in G.nodes(data=True)]

# Visualize the network with gender-based colors
pos = nx.spring_layout(G)
nx.draw(G, pos, node_size=20, node_color=node_colors, with_labels=False)
plt.title("Speed Dating Network with Gender-Based Colors")
plt.show()
```

```python
from matplotlib.cm import ScalarMappable
from matplotlib.colors import Normalize

# Calculate and visualize degree centrality
degree_centrality = nx.degree_centrality(G)
plt.figure(figsize=(10, 6))
degree_centrality_values = [degree_centrality[node] for node in G.nodes()]
cmap = plt.get_cmap('Reds')
normalize = Normalize(vmin=min(degree_centrality_values), vmax=max(degree_centrality_values))
scalar_mappable = ScalarMappable(cmap=cmap, norm=normalize)
scalar_mappable.set_array([])
ax = plt.gca()  # Get the current axes
nx.draw(G, pos, node_size=20, node_color=degree_centrality_values, cmap=cmap, with_labels=False)
plt.title("Degree Centrality")
# Specify the ax parameter for colorbar
colorbar = plt.colorbar(scalar_mappable, ax=ax, label="Degree Centrality")
plt.show()
```

## Analyzing the network

```python
# calculate Degree Centrality
degree_centrality = nx.degree_centrality(G)
# sort Degree Centrality values in descending order
sorted_nodes = sorted(degree_centrality.items(), key=lambda x: x[1], reverse=True)

# calculate the top 30% nodes
top_percent_count = int(0.30 * len(G))

# select the top nodes with the highest Degree Centrality
top_degree_nodes = [node for node, centrality in sorted_nodes[:top_percent_count]]

# Extract the attributes of these nodes and remove the duplicate iid
top_degree_nodes_data = df[df['iid'].isin(top_degree_nodes)].drop_duplicates(subset=['iid'])
average_age = top_degree_nodes_data['age'].mean()
median_age = top_degree_nodes_data['age'].median()
max_age = top_degree_nodes_data['age'].max()
min_age = top_degree_nodes_data['age'].min()
gender_distribution = top_degree_nodes_data['gender'].value_counts()
race_distribution = top_degree_nodes_data['race'].value_counts()
career_distribution = top_degree_nodes_data['career_c'].value_counts()
field_distribution = top_degree_nodes_data['field_cd'].value_counts()

# Analyze basic attributes
samerace_distribution = top_degree_nodes_data['samerace'].value_counts()
imprace_distribution = top_degree_nodes_data['imprace'].value_counts()
imprelig_distribution = top_degree_nodes_data['imprelig'].value_counts()

# Create a new figure and a grid of subplots
fig, axes = plt.subplots(2, 4, figsize=(30, 15))

# Plot Age Distribution
axes[0, 0].hist(top_degree_nodes_data['age'].sort_index(), bins=20, color='skyblue')
axes[0, 0].set_title('Age Distribution')
axes[0, 0].set_xlabel('Age')
axes[0, 0].set_ylabel('Count')

# Plot Gender Distribution
gender_distribution.sort_index().plot(kind='bar', color='lightcoral', ax=axes[0, 1])
axes[0, 1].set_title('Gender Distribution')
axes[0, 1].set_xlabel('Gender')
axes[0, 1].set_ylabel('Count')

# Plot Race Distribution
race_distribution.sort_index().plot(kind='pie', autopct='%1.1f%%',
                                    colors=['lightgreen', 'lightblue', 'lightyellow', 'lightpink'],
                                    ax=axes[0, 2])
axes[0, 2].set_title('Race Distribution')
axes[0, 2].set_ylabel('')

# Plot Career Distribution
career_distribution.sort_index().plot(kind='bar', color='lightcoral', ax=axes[0, 3])
axes[0, 3].set_title('Career Distribution')
axes[0, 3].set_xlabel('Career Code')
axes[0, 3].set_ylabel('Count')

# Plot Field Distribution
field_distribution.sort_index().plot(kind='bar', color='lightblue', ax=axes[1, 0])
axes[1, 0].set_title('Field Distribution')
axes[1, 0].set_xlabel('Field Code')
axes[1, 0].set_ylabel('Count')

# Plot Distribution of Same Race
samerace_distribution.sort_index().plot(kind='bar', color='skyblue', ax=axes[1, 1])
axes[1, 1].set_title('Distribution of Same Race')
axes[1, 1].set_xlabel('Same Race')
axes[1, 1].set_ylabel('Count')

plt.figure(figsize=(10, 6))
imprace_distribution.sort_index().plot(kind='bar', color='lightcoral', ax=axes[1, 2])
axes[1, 2].set_title('Distribution of Preference for Dating Partner\'s Race')
axes[1, 2].set_xlabel('Preference Level')
axes[1, 2].set_ylabel('Count')

plt.figure(figsize=(10, 6))
imprelig_distribution.sort_index().plot(kind='bar', color='lightcoral', ax=axes[1, 3])
axes[1, 3].set_title('Distribution of Preference for Dating Partner\'s Religion')
axes[1, 3].set_xlabel('Preference Level')
axes[1, 3].set_ylabel('Count')

# Adjust layout and display the combined plot
plt.tight_layout(h_pad=5, w_pad=5)
plt.show()
```
